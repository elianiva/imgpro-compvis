````python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io, transform
from PIL import Image
import matplotlib.pylab as plt
````

````python
urls = [
    "https://car-images.bauersecure.com/wp-images/12120/480x270/gtrnismo_12.jpg",
    "https://car-images.bauersecure.com/wp-images/12120/480x270/gtrnismo_07.jpg",
    "https://car-images.bauersecure.com/wp-images/12120/480x270/gtrnismo_01.jpg"
]

for url in urls:
    image = io.imread(url)
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    final_frame = cv.hconcat((image, image_2))
    cv2_imshow(final_frame)
    print("\n")
````

````python
height, width = image_2.shape[:2]
print("Image Resolution: width x height = ", width, " x ", height)
cv2_imshow(image_2)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3[int(height/2)] = [[255, 255, 255] for _ in range(width)]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
print('\n')

image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for i in range(height):
    image_4[i][int(width/2)] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame)
print('\n')

# diagonal line
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
width, height = image_5.shape[1], image_5.shape[0]
for i in range(height):
  image_5[i][int(i * (width / height))] = [255, 255, 255]
  image_5[height - i - 1][int(i * (width / height))] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
print('\n')

# limited line
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for i in range(100, 200):
  image_6[int(height/2)+40][i] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_6))
cv2_imshow(final_frame)
print('\n')

# square
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for i in range(20, 180):
  for j in range(20):
    image_7[height-20-j][i] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_7))
cv2_imshow(final_frame)
print('\n')
````

````python
from random import randint

id_card = io.imread("https://umsu.ac.id/artikel/wp-content/uploads/2023/11/cara-mudah-cek-ktp-asli-atau-palsu-718x375.jpeg")
id_card = cv.cvtColor(id_card, cv.COLOR_BGR2RGB)
# cv2_imshow(id_card)

height, width = id_card.shape[:2]
print("Image Resolution: width x height = ", width, " x ", height)
print('\n')

def draw_square(width, height, dx, dy):
  for i in range(width):
    for j in range(height):
      id_card[dy+j][dx+i] = [randint(0, 255), randint(0, 255), randint(0, 255)]

draw_square(240, 18, 200, 80)
draw_square(120, 16, 225, 111)
draw_square(150, 16, 225, 128)
draw_square(240, 16, 225, 145)
draw_square(160, 16, 225, 162)
draw_square(110, 16, 225, 180)
draw_square(120, 16, 225, 196)
draw_square(100, 16, 225, 212)
draw_square(100, 16, 225, 232)
draw_square(60, 16, 225, 249)
draw_square(130, 16, 225, 266)
draw_square(40, 16, 225, 284)
draw_square(80, 16, 225, 302)
draw_square(120, 30, 495, 280)
draw_square(120, 160, 500, 100)
draw_square(80, 50, 520, 315)
draw_square(250, 50, 240, 4)
cv2_imshow(id_card)
````
